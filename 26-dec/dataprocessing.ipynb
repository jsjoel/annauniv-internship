```python
ohe = OneHotEncoder(sparse_output=False, drop= None)

encoded_geography = ohe.fit_transform(df[['Geography']])

ohe_columns = ohe.get_feature_names_out(['Geography'])
encoded_df = pd.DataFrame(encoded_geography, columns=ohe_columns)

df = df.drop(columns=['Geography'])
for i, col in enumerate(encoded_df.columns):
    df.insert(loc=i+1, column = col, value = encoded_df[col])
```

Outlier Detection

```python
df_1 = df.copy()
df_2 = df.copy()
```

```python
df.boxplot(column='CreditScore')
plt.show()
```

```python
Q1 = df_1['CreditScore'].quantile(0.25)
Q3 = df_1['CreditScore'].quantile(0.75)
IQR = Q3 - Q1
upper_limit = Q3 + 1.5*IQR
lower_limit = Q1 - 1.5*IQR

upper_array = np.where(df_1.CreditScore >= upper_limit)[0]
lower_array = np.where(df_1.CreditScore <= lower_limit)[0]

df_1.drop(index=upper_array, inplace=True )
df_1.drop(index = lower_array, inplace=True)
```

```python
df_1.boxplot(column='CreditScore')
plt.show()
```
